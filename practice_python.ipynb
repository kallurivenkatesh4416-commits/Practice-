```python
def factorial(n: int) -> int:
    if not isinstance(n, int):
        raise TypeError("n must be an int")
    if n < 0:
        raise ValueError("n must be >= 0")
    fact = 1
    i = 2
    while i <= n:
        fact *= i
        i += 1
    return fact

# tests
for t in [5, 0, 1]:
    print(t, factorial(t))

try:
    factorial(-3)
except ValueError as e:
    print("error:", e)
```

```python
def summarize_iter(*num, op = 'mean'):
    if not num:
        raise ValueError('No numbers provided')

    if op == 'mean':
        return sum(num)/len(num)
    elif op == 'sum':
        return sum(num)
    elif op == 'max':
        return max(num)
    elif op == 'min':
        return min(num)
    elif op == 'multiply':
        prod = 1
        for x in num:
            prod *= x
        return prod
           
                        
    else:
        raise ValueError(f"Unsupported op: {op!r}. Use 'mean' or 'sum'or 'min' or 'max' and 'multiply'.")

summarize_iter(1,2,3,4,5, op = 'multiply')
```
